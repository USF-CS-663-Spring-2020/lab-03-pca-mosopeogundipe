```python
# Use sklearn to standardise (specifically: normalise) the data so that 
# the data has a normal distribution -- mean = 0; stdev = 1
from sklearn.preprocessing import StandardScaler
import numpy as np

x = df.values
x = StandardScaler().fit_transform(x)

np.mean(x),np.std(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_breast = pd.DataFrame(x,columns=feat_cols)
normalised_breast.tail()
normalised_breast
```

```python
from sklearn.decomposition import PCA

pca_handwriting = PCA(n_components=4)
principalComponents_handwriting = pca_handwriting.fit_transform(x)

principal_handwriting_df = pd.DataFrame(data = principalComponents_handwriting
             , columns = ['principal component 1', 'principal component 2', 'principal component 3', 'principal component 4'])
```

### Variation Explained per principal component

It makes sense that the a smaller percentage of variation is explained per principal component for this dataset, when compared with the WDBC dataset because:

* This dataset has many more features (columns) than the WDBC dataset. It has 5000 columns compared with WDBC's 30 columns.
* The columns in this dataset aren't as closely correlated as those in the WDBC dataset. This makes sense because the probability of correlation of columns would decrease with an increase in number of columns.

```python
# The Explained variation per principal component is given below
# 3.16% is explained by PC1; 2.57% by PC2; 1.99% by PC3, 1.77% by PC4:
print('Explained variation per principal component: {}'.format(pca_handwriting.explained_variance_ratio_))
```

### How many principal components do we need to explain the variance in data?

We need all the principal components in this dataset to express all the variance in the data. As can be seen from the cell below, the variance explained only becomes zero at the last column. This also goes to reinforce what I explained in the previous section that the columns are not closely correlated.

```python
# How many principal components do we need to explain the variance in data?
pca = PCA().fit(x)

evr = pca.explained_variance_ratio_
print('\n'.join('{:2d}: {:0.4f}'.format(*k) for k in enumerate(evr)))
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Total Explained Variance');
plt.title('Number of Components Needed to Explain all the Variance', fontSize=10)
```
